<h3 style="color: #2e6c80;">2.3 Limpiando la base de datos</h3>

* Primero, importaremos nuestras bases de datos y les daremos un nombre para diferenciarlas.

```python
base_origen = pd.read_csv('owid-covid-data.csv')
```

```python
base_origen
```

* Con el comando `shape` obtenemos el número de filas y columnas de la base de datos. Observamos que tenemos 60 columnas que corresponden a las distintas variables consideradas en la base de datos. Sin embargo, solo nos quedaremos con las variables más relevantes para el análisis estadístico. Estas serán el continente, país, fecha, casos totales de infectados, nuevos casos de infectados, total de muertes, y nuevas muertes.

```python
base_origen.shape
```

```python
base_origen = base_origen.loc[:,['continent', 'location', 'date', 'total_cases', 
                                 'new_cases', 'total_deaths', 'new_deaths']]
```

* Un paso importante que se debe de realizar es comprobar que no hayan datos missing (NaN). Para ello, se emplea el comando `df.isna().sum()`.

```python
base_origen.isna().sum()
```

* Observamos que hay muchos datos missing en las columnas, por lo cual, los rellenaremos con 0. Más que nada, este problema ocurre porque el impacto del virus no sucedió al mismo tiempo en todos los países. Así que según las fechas, hay países en que el virus infectó mucho después respecto al inicio de la pandemia.

```python
base_origen['total_cases'] = base_origen['total_cases'].fillna(0)
base_origen['new_cases'] = base_origen['new_cases'].fillna(0)
base_origen['total_deaths'] = base_origen['total_deaths'].fillna(0)
base_origen['new_deaths'] = base_origen['new_deaths'].fillna(0)
```

* Ahora, comprobaremos qué tipo de datos son nuestras variables para no tener inconvenientes más adelante al realizar los gráficos.

```python
base_origen.dtypes
```

* Tenemos un problema con `date`, ya que es tipo object, lo cual es incorrecto dado que representa a una fecha. Lo solucionamos transformando su formato a fecha.

```python
base_origen['date'] = pd.to_datetime(base_origen['date'])
base_origen.dtypes
```

* Le cambiamos el nombre a las columnas

```python
base_origen.columns = ['Continente', 'País', 'fecha', 'casos totales', 'nuevos casos', 'muertes totales', 'nuevas muertes']
base_origen
```

* Ahora limpiaremos la base de datos sobre la vacunación a nivel mundial.

```python
vacunacion = pd.read_csv('owid-covid-data 1.csv')
```

```python
vacunacion
```

* Seleccionamos las variables relevantes

```python
vacunacion = vacunacion.loc[:,['continent', 'location', 'date', 'total_vaccinations', 'population', 'new_cases']]
```

* Dado que se desea analizar el total de personas vacunadas hasta el 24 de junio de 2021, se filtra por fecha la base de de datos de la vacunación.

```python
vacunas_mundial = vacunacion[vacunacion['date']=='2021-06-24']
```

* Finalmente, cambiamos nombres de las columnas y realizamos un groupby por continentes. Aquí también se cambiarán los nombres de las columnas. Además, como se observa que el total de vacunados está en notación científica, se procede a cambiarle el formato.

```python
vacunas_mundial.columns = ['Continente', 'País', 'fecha', 'vacunados_totales', 'población', 'nuevos_casos']
pd.options.display.float_format = '{:.4f}'.format
vacunas_continente = vacunas_mundial.groupby(by = ['Continente']).sum().reset_index()
vacunas_continente
```

<h2 style="color: #2e6c80;">3. Análisis Gráfico</h2>
<h3 style="color: #2e6c80;">3.1 Contagios en los principales países</h3>

* Primero, seleccionaremos a los principales países para el análisis.

```python
paises = ['Russia', 'United States', 'France', 'India', 'Brazil'] 
graf1 = base_origen[base_origen['País'].isin(paises)]
```

```python
# Primero creamos una variable para realizar la selección inteactiva que tendrá el gráfico de líneas
nearest = alt.selection(type='single', nearest=True, on='mouseover',
                        fields=['fecha'], empty='none')

# Ahora creamos el gráfico de líneas del número de infectados por COVID entre el 2020 y 2021 para los países elegidos
plot1 = alt.Chart(graf1).mark_line(size=2).encode(
    alt.X('fecha:T',
         axis=alt.Axis(title='Meses de Pandemia por COVID-19', titleFontSize=14, grid=False)),
    alt.Y('casos totales:Q',
         axis=alt.Axis(title='Número de Infectados', titleFontSize=15, grid=False)),
     color='País:N',
     opacity=alt.value(0.5)
)
# Está línea de código es para crear los selectores que interactuarán a lo largo del gráfico mientras movamos el cursor

selectors = alt.Chart(graf1).mark_point().encode(
    x='fecha:T',
    opacity=alt.value(0),
).add_selection(
    nearest
)
    
    
# Aquí creamos los puntos que se resaltarán en la selección
points = plot1.mark_point().encode(
    opacity=alt.condition(nearest, alt.value(1), alt.value(0))
)

# Y estas son las etiquetas que se mostrarán con los puntos, en este caso, nos interesa saber 
# el número de casos totales de infectados
text = plot1.mark_text(align='left', dx=5, dy=-5).encode(
    text=alt.condition(nearest, 'casos totales:Q', alt.value(' '))
)

# Añadimos una regla que nos muetres cómo nos desplazamos por las curvas a lo largo de las fechas
rules = alt.Chart(graf1).mark_rule(color='gray').encode(
    x='fecha:T',
).transform_filter(
    nearest
)
# Ahora que tenemos todas las capas necesarias, las unimos en una sola llamada "final1". 
#Además, añadimos el título al gráfico
    
final1 = alt.layer(
    plot1, selectors, points, rules, text
).properties(title=alt.TitleParams(
    "Número de casos de infectados por COVID-19 (2020-2021)",
    color='black',
fontSize = 25,
dy = -30), width=750, height=700)

# Finalmente, dado que Altair no posee una sintaxis propia para añadir pie de página,
# utilizamos la instrucción "concat" para crear un segundo título que servirá como pie
# de página para colocar la fuente del gráfico respectiva 
    
alt.concat(final1,
    title=alt.TitleParams(
        'Fuente: Elaboración propia con datos de Our World in Data ',
        color='gray',
        baseline='bottom',
        orient='bottom',
        anchor='start',
        dy = 20,
        fontSize= 10
    )
)
```

* En el gráfico se puede apreciar cómo han existido momentos en que el número de contagios se suavizaba, para luego incrementarse en gran medida. Un claro ejemplo es el periodo de noviembre - diciembre de 2020, donde los cinco países presentan una mayor pendiente. Una explicación para el aumento de contagios fue la cercanía de las fiestas navideñas y fin de año. Otro periodo crítico fue el de marzo-abril del presente año. Se puede apreciar cómo las curvas de Brasil, Francia, India y los Estados Unidos se elevan en poco tiempo.

<h3 style="color: #2e6c80;">3.2 Número de muertes en América Latina</h3>

* Ahora se realizará un gráfico de las muertes diarias sucedidas en los principales países de América Latina durante el último mes de junio

```python
# Ahora seleccionamos otro grupo de país y filtramos por fecha para solo tener datos de junio de 2021
seleccion2 = ['Peru', 'Mexico', 'Chile', 'Argentina', 'Colombia', 'Ecuador', 'Panama'] 
graf2 = base_origen[base_origen['País'].isin(seleccion2)]
graf2 = graf2[graf2.fecha >'2021-05-31']
```

```python
# El gráfico será uno de círculos combinado con uno de barras. Para ello, primero definimos la variable de intervalo
# que servirá para interactuar a través de selecciones que se podrán hacer con el mouse

interval = alt.selection_interval()

# Gráfico de círculos
circulos = alt.Chart(graf2).mark_circle().encode(
    alt.X('monthdate(fecha):O',
         axis=alt.Axis(title='Días de junio en pandemia por COVID-19', titleFontSize=12)),
    alt.Y('País',
         axis=alt.Axis(title='Países de América Latina', titleFontSize=14)),
    color=alt.condition(interval, 'País', alt.value('lightgray')),
    tooltip=['País','nuevas muertes'],
    size=alt.Size('nuevas muertes:Q',
        scale=alt.Scale(range=[0, 3000]),
        legend=alt.Legend(title='Muertes diarias')
    ) 
).properties(title=alt.TitleParams(
    "Número de muertes diarias durante junio de 2021",
    color='black',
    fontSize = 25,
    dy = -30),width=800,
    height=300,
    selection=interval).interactive()

#  Pie de página
alt.concat(circulos,
    title=alt.TitleParams(
        'Fuente: Elaboración propia con datos de Our World in Data ',
        color='gray',
        baseline='bottom',
        orient='bottom',
        anchor='start',
        dy = 20,
        fontSize= 10
    )
)
```

* En este gráfico se muestra el número de muertes diarias durante el mes de junio para siete países en latinoamérica. Se puede observar cómo las muertes se han mantenido bajas a lo largo de los días. El país con menos muertes ha sido Panamá, seguido de Ecuador y Chile. Otro dato resaltante son los días en los que no se ha regristrado ninguna muerte en países como Perú, Panamá y Ecuador. Sin embargo, Argentina y Colombia han mantenido un alto número de muertes a lo largo del mes.

<h3 style="color: #2e6c80;">3.2 Avance de la vacunación a nivel mundial</h3>

```python
# Ahora queremos un gráfico que muestra el avance de la vacunación a nivel mundial hasta el 24 de junio de 2021.
# Este gráfico debe de mostrar el porcentaje de vacunados totales por continente y compararlo con el total
# de población

# Primero, creamos el plot3 que tendrá a la población ya vacunada
plot3 = alt.Chart(vacunas_continente).transform_joinaggregate(
    TotalTime='sum(población)',
).transform_calculate(
    PercentOfTotal="datum.vacunados_totales / datum.TotalTime"
).mark_bar().encode(
    alt.Y('PercentOfTotal:Q', axis=alt.Axis(title='Porcentaje de la población mundial', titleFontSize=12, format='.0%')),
    alt.X('Continente:N',
          axis=alt.Axis(title='Continentes', titleFontSize=12)), 
    color = 'Continente:N'
).properties(width=300,
    height=400)

# Ahora, creamos el plot4 que tiene a la población por continente en porcentaje respecto al total poblacional del mundo
plot4 = alt.Chart(vacunas_continente).transform_joinaggregate(
    TotalTime='sum(población)',
).transform_calculate(
    PercentOfTotal="datum.población / datum.TotalTime"
).mark_bar(opacity=0.5).encode(
    alt.Y('PercentOfTotal:Q', axis=alt.Axis(format='.0%')),
    x='Continente:N',
    color = alt.Color('Continente:N', scale=alt.Scale(scheme='accent')),
    tooltip=['Continente', 'vacunados_totales']
).properties(title=alt.TitleParams(
    "Porcentaje de vacunados respecto a la población mundial según continente",
    color='black',
    fontSize = 15,
    dy = -30),width=300,
    height=400).interactive()

#  Finalmente, creamos el pie de página donde va la fuente
alt.concat(plot3 + plot4,
    title=alt.TitleParams(
        'Fuente: Elaboración propia con datos de Our World in Data ',
        color='gray',
        baseline='bottom',
        orient='bottom',
        anchor='start',
        dy = 20,
        fontSize= 10
    )
)


```

* Finalmente, este tercer gráfico representa el porcentaje de la población mundial total y la vacunada. Como se observa, Asia que representa el 60% de la población mundial, ya lleva vacunada más de un 20%. Mientras que Europa lleva vacunada casi la mitad de su población. Norte América lleva la delantera en vacunados respecto a su total. Sin embargo, África es el continente con menor número de vacunados, pese a que es el segundo continente con mayor población.
